```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
plt.rcParams['figure.figsize'] = 12,8
```

```python
df = pd.read_csv('NCHS_-_Death_rates_and_life_expectancy_at_birth.csv')
```

```python
df.head()
```

```python
df = df.set_index('Year')
```

```python
df.head()
```

```python
df.columns
```

```python
df.columns = ['Race', 'Gender', 'Avg_Life_Exp', 'Age_Adj_Death_Rate']
```

```python
df.head()
```

```python
df.tail()
```

```python
df_white_male = df[(df.Gender=='Male') & (df.Race=='White')]
df_white_female = df[(df.Gender=='Female') & (df.Race=='White')]
```

```python
df_black_male= df[(df.Gender=='Male') & (df.Race=='Black')]
df_black_female=df[(df.Gender=='Female') & (df.Race=='Black')]
```

```python
plt.plot(df_white_male['Age_Adj_Death_Rate'], c='Blue', marker='o', ms=7, ls='--')
plt.plot(df_black_male['Age_Adj_Death_Rate'], c='Red', ms=7, ls='-', marker='^')
plt.plot(df_white_female['Age_Adj_Death_Rate'], c='Green', ms=7, ls='--', marker='o')
plt.plot(df_black_female['Age_Adj_Death_Rate'], c='Orange', ms=7, ls='', marker='^')
```

```python
plt.plot(df_white_male['Avg_Life_Exp'], c='Blue', marker='o', ls='--', ms=7)
plt.plot(df_black_male['Avg_Life_Exp'], c='Red', marker='s', ls='', ms=7)
plt.plot(df_white_female['Avg_Life_Exp'], c='Green', ms=7, ls='--', marker='o')
plt.plot(df_black_female['Avg_Life_Exp'], c='Orange', ms=7, ls='', marker='^')
```
